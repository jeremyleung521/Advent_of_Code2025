### Lafazar's Solution

```python
with open("Day9_input.txt") as file:
    corners = [[int(s) for s in line.split(",")] for line in file.read().split("\n")]

max((abs(x2-x1)+1)*(abs(y2-y1)+1) for x1, y1 in corners for x2, y2 in corners)
```

```python
areas = sorted([((abs(x2-x1)+1)*(abs(y2-y1)+1), (min(x1, x2), min(y1, y2), max(x1, x2), max(y1, y2)))
                for i, (x1, y1) in enumerate(corners) for x2, y2 in corners[i+1:]], reverse=True)
lines = sorted([(abs(x2-x1)+abs(y2-y1), (min(x1, x2), min(y1, y2), max(x1, x2), max(y1, y2)))
                for (x1, y1), (x2, y2) in zip(corners, corners[1:]+[corners[0]])], reverse=True)
for area, (cx1, cy1, cx2, cy2) in areas:
    if not any(lx1 < cx2 and ly1 < cy2 and lx2 > cx1 and ly2 > cy1 for length, (lx1, ly1, lx2, ly2) in lines):
        break
area
```
